# 1. Inner Join

```
spark.sql("""select Ename,Job,sal,emp.deptno,dname,Loc 
             from emp inner join dept
             on  emp.deptno=dept.deptno
          """).show()
```

# 2. Left Outer Join

```
spark.sql("""select Ename,Job,sal,emp.deptno,dname,Loc 
             from emp left outer join dept
             on  emp.deptno=dept.deptno
          """).show()
```

# 3. Right Outer Join

```
spark.sql("""select Ename,Job,sal,emp.deptno,dname,Loc 
             from emp right outer join dept
             on  emp.deptno=dept.deptno
          """).show()
```

# 4. Full Outer Join

```
spark.sql("""select Ename,Job,sal,emp.deptno,dname,Loc 
             from emp full outer join dept
             on  emp.deptno=dept.deptno
          """).show()
```

# 5. Non Equi Join

```
spark.sql("""select Ename,Job,sal,grade
             from emp,salgrade
             where sal>=losal and sal<=hisal
          """).show()
```

# 6. Cross Join

```
spark.sql("""select Ename,Job,sal,grade
             from emp,salgrade
             where sal>=losal and sal<=hisal
          """).show()
```

# Group By

```
spark.sql("""select dname, sum(sal) as total_sal from emp,dept
          where emp.deptno=dept.deptno
          group by dname""").show()
```
